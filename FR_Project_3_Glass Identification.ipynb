```python
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve,roc_auc_score,classification_report
```

```python
x= df.drop(columns = ['Type_Of_Glass'])
y = df['Type_Of_Glass']
```

```python
scaler= StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
vif = pd.DataFrame()
vif['vif']=[variance_inflation_factor(x_scaled,i)for i in range(x_scaled.shape[1])]
vif['Features']=x.columns


vif
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state =58)
```

```python
def metric_score(clf,x_train,x_test,y_train,y_test,train = True):
    if train:
        y_pred =clf.predict(x_train)
        
        print ("\n..................Train Result.....................")
        
        print(f"Accuracy score: {accuracy_score(y_train,y_pred)*100:.2f}%")
              
    elif train == False:
        pred =clf.predict(x_test)
        print("\n..................Test Result.......................")
        print(f"Accuracy score : {accuracy_score(y_test,pred)*100:.2f}%")
        print("\n\n Test Classification report  \n",classification_report(y_test,pred,digits =2))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
metric_score(clf,x_train, x_test,y_train,y_test,train = True)

metric_score(clf,x_train, x_test,y_train,y_test,train = False)
```

```python
y_pred=clf.predict(x_test)
```

```python
conf_mat=confusion_matrix(y_test,y_pred)
conf_mat
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
grid_param = {
            'criterion':['gini','entropy'],
             'max_depth':range(20,25),       
              'min_samples_leaf': range(2,3),  
             'min_samples_split':range(3,4),   
              'max_leaf_nodes':range(20,30)}
```

```python
grid_search = GridSearchCV(estimator = clf,
                           param_grid= grid_param,cv=5)
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_parameters = grid_search.best_params_

print(best_parameters)
```

```python
clf =RandomForestClassifier(criterion='gini', max_depth = 21, max_leaf_nodes=27, min_samples_leaf= 2,min_samples_split=3)

clf.fit(x_train,y_train)
```

```python
metric_score(clf,x_train, x_test,y_train,y_test,train = True)

metric_score(clf,x_train, x_test,y_train,y_test,train = False)
```
